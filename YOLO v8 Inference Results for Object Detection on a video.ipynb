## 1. YoloV8 on a video for object detection

```python
videofile = "paris.mp4"
```

```python
!ls $videofile -lh
```

```python
minconf = 0.3
yolov8model = "yolov8l.pt"
```

```python
!yolo task=detect mode=predict model=$yolov8model conf=$minconf source=$videofile line_thickness=1
```

```python
!ls runs/detect/predict/*.mp4 -lh
```

```python
video_file_result = "runs/detect/predict/" + videofile
video_compressed_file = "object_detection_" + videofile
```

```python
from moviepy.editor import *

clip = VideoFileClip(video_file_result)
clip.write_videofile(video_compressed_file)
```
